```python
import pandas as pd
import numpy as np

# Load the dataset into a DataFrame
df = pd.read_csv(r"C:\Users\shiva\Downloads\EDA1\EDA1\Cardiotocographic.csv")

# Handle missing values
df.fillna(df.mean(), inplace=True)

# Convert data types if necessary
# No conversion needed in this case as data types are assumed to be correct

# Detect and treat outliers if necessary
from scipy.stats import zscore
z_scores = zscore(df)
abs_z_scores = np.abs(z_scores)
filtered_entries = (abs_z_scores < 3).all(axis=1)
df = df[filtered_entries]

# Save the cleaned dataset
df.to_csv("cleaned_cardiographic.csv", index=False)
```

```python
# Statistical summary for each variable
stat_summary = df.describe()
print(stat_summary)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histograms for numerical variables
num_vars = ['LB', 'AC', 'FM', 'UC', 'DL', 'DS', 'DP', 'ASTV', 'MSTV', 'ALTV', 'MLTV']
for var in num_vars:
    plt.figure(figsize=(8, 6))
    sns.histplot(df[var], bins=20, kde=True)
    plt.title(f'Histogram of {var}')
    plt.xlabel(var)
    plt.ylabel('Frequency')
    plt.show()

# Bar charts for categorical variables (if any)
# No categorical variables assumed in this dataset

# Scatter plot matrix or correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
